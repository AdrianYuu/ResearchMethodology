```python
import numpy as np
import matplotlib.pyplot as plt

# Classifier data
classifiers = ['Naïve Bayes', 'Random Forest', 'Support Vector Machine (SVM)']
accuracy = [0.66, 0.72, 0.73]
precision = [0.77, 0.74, 0.77]
recall = [0.59, 0.79, 0.78]
f1_score = [0.67, 0.77, 0.77]

# Set the width of the bars
bar_width = 0.15

# Set the x locations for the groups
index = np.arange(len(classifiers))

# Plotting the bars
plt.figure(figsize=(10, 5))
plt.bar(index - 1.5 * bar_width, accuracy, bar_width, label='Accuracy')
plt.bar(index - 0.5 * bar_width, precision, bar_width, label='Precision')
plt.bar(index + 0.5 * bar_width, recall, bar_width, label='Recall')
plt.bar(index + 1.5 * bar_width, f1_score, bar_width, label='F1-Score')

# Adding labels and title
plt.xlabel('Classifier')
plt.ylabel('Score')
plt.title('Classifier Metrics')
plt.xticks(index, classifiers)

# Move the legend outside the plot
plt.legend(loc='upper right', bbox_to_anchor=(1.17, 1))

plt.grid(True)
plt.tight_layout()
plt.show()


```
